## Bentley auth using the Evo Python SDK

This notebook shows the two main SDK authentication flows:

- Native, SPA, or Web apps that use the authorization code flow.
- Service apps that use the client credentials flow.

If you need direct API examples instead, check out the `api-examples.ipynb` notebook.

### Native, SPA, or Web app auth

Create a **native**, **SPA**, or **web app** in the [iTwin Developer Portal](https://developer.bentley.com/register/?product=seequent-evo). This gives you the client ID and redirect URL that the SDK needs in the next cell to start the sign-in flow.

```python
# Enter your client ID and callback URL.

client_id = "<your-client-id>"
redirect_url = "<your-redirect-url>"

from evo.notebooks import ServiceManagerWidget

# Start the authorization code login flow in the notebook UI and create a manager
# that stores the authenticated Evo session after sign-in completes.
manager = await ServiceManagerWidget.with_auth_code(
    redirect_url=redirect_url,
    client_id=client_id,
).login()
```

```python
# Read the authenticated environment from the session manager and print the
# core identifiers you can use in later SDK or API examples.
environment = manager.get_environment()
print(f"Hub URL: {environment.hub_url}")
print(f"Org ID: {environment.org_id}")
print(f"Workspace ID: {environment.workspace_id}")
```

### Service app auth

Create a **service app** in the [iTwin Developer Portal](https://developer.bentley.com/register/?product=seequent-evo). This gives you the client ID and client secret that the SDK needs in the next cell to start the sign-in flow.

NOTE: Delete the client secret from this notebook before committing it to source control or sharing it with others.

```python
from evo.aio import AioTransport
from evo.oauth import ClientCredentialsAuthorizer, OAuthConnector

# Enter your client ID and client secret from the Service application you created
# in the iTwin Developer Portal. The Evo scopes are handled automatically by the SDK.
client_id = "<your-service-app-client-id>"
client_secret = "<your-service-app-client-secret>"

# Set a user agent string so requests made during authentication and later API calls
# can be identified by the remote service.
user_agent = "Evo demo app/1.0"

# Build an OAuth connector that knows how to talk to the token service using your
# client credentials and an HTTP transport configured with the user agent above.
# Then wrap it in a ClientCredentialsAuthorizer, which requests and refreshes access
# tokens for this service application when headers are needed.
authorizer = ClientCredentialsAuthorizer(
    oauth_connector=OAuthConnector(
        transport=AioTransport(user_agent=user_agent),
        client_id=client_id,
        client_secret=client_secret,
    ),
)

# Request the default headers from the authorizer. This triggers token acquisition if
# needed and returns headers that include the Authorization bearer token for use in
# later SDK or API calls.
print(await authorizer.get_default_headers())
```
