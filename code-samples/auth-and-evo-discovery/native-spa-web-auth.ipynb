## Bentley auth using native, SPA or web app credentials

This notebook shows how to sign in with an application that uses the authorization code flow and then create an Evo SDK session from the returned credentials.

Create a Native, SPA, or Web application in the [iTwin Developer Portal](https://developer.bentley.com/register/?product=seequent-evo). This gives you the client ID and redirect URL that the SDK needs in the next cell to start the sign-in flow.

If you need an API example instead, check out the `api-examples.ipynb` notebook. This notebook focuses on the SDK-driven login path.

```python
# Enter your client ID and callback URL.

client_id = "<your-client-id>"
redirect_url = "<your-redirect-url>"

from evo.notebooks import ServiceManagerWidget

# Start the authorization code login flow in the notebook UI and create a manager
# that stores the authenticated Evo session after sign-in completes.
manager = await ServiceManagerWidget.with_auth_code(
    redirect_url=redirect_url,
    client_id=client_id,
).login()
```

```python
# Read the authenticated environment from the session manager and print the
# core identifiers you can use in later SDK or API examples.
environment = manager.get_environment()
print(f"Hub URL: {environment.hub_url}")
print(f"Org ID: {environment.org_id}")
print(f"Workspace ID: {environment.workspace_id}")
```
