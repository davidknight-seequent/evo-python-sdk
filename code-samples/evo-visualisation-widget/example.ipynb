# Evo Visualisation Widget — Example

Render Evo geoscience objects (surfaces, pointsets, downhole data, ...) in a notebook,
alongside 3D axes and numbered tick labels. Block models and Central objects are out of scope.

Make sure the `evo_viz` package sits next to this notebook (or is installed), and that you have
installed the dependencies from `requirements.txt`.

```python
from evo_viz import (
    login,
    list_visualizable_objects,
    download_tileset_bundle,
    EvoObjectViewer,
)

# Replace with your registered Evo application credentials.
CLIENT_ID = "<your-client-id>"
REDIRECT_URL = "<your-redirect-url>"  # e.g. http://localhost:3000/signin

manager = await login(client_id=CLIENT_ID, redirect_url=REDIRECT_URL)
```

## Browse renderable objects

This lists only the schema types the Visualisation service supports, with block models filtered out.

```python
objects = await list_visualizable_objects(manager)
for i, obj in enumerate(objects):
    print(f"{i:>3}  {obj.name}  ({obj.schema})")
```

## Download a tileset and view it

`download_tileset_bundle` fetches the OGC 3D Tiles tileset (polling while it is generated) and
downloads every referenced payload through the authenticated connector.

```python
target = objects[0]
bundle = await download_tileset_bundle(manager, target.object_id, name=target.name)

viewer = EvoObjectViewer(axis_labels=["Easting", "Northing", "Elevation"])
viewer.add_bundle(bundle)
viewer
```

## Add more objects to the same scene

Every bundle is independent, so you can layer several objects together.

```python
another = await download_tileset_bundle(manager, objects[1].object_id, name=objects[1].name)
viewer.add_bundle(another)
```
